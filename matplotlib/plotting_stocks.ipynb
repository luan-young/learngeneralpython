```python
#aapl_df = yf.download('AAPL', period='10y')
#aapl_df = yf.download('AAPL', period='max')
aapl_df = yf.download('AAPL', start='2010-01-01', end='2015-01-01')
msft_df = yf.download('MSFT', start='2010-01-01', end='2015-01-01')
goog_df = yf.download('GOOG', start='2010-01-01', end='2015-01-01')
```

```python
aapl_df.head()
```

```python
msft_df.head()
```

```python
goog_df.head()
```

```python
aapl_df['Normalized Close'] = aapl_df['Adj Close'] / aapl_df['Adj Close'][0]
msft_df['Normalized Close'] = msft_df['Adj Close'] / msft_df['Adj Close'][0]
goog_df['Normalized Close'] = goog_df['Adj Close'] / goog_df['Adj Close'][0]
```

```python
steves_death = datetime64('2011-10-05')

style.use('fivethirtyeight')

fig, ax = plt.subplots(figsize=(10, 6))
ax.plot('Normalized Close', label='Apple', data=aapl_df)
ax.plot('Normalized Close', label='Microsoft', data=msft_df)
ax.plot('Normalized Close', label='Google', data=goog_df)
ax.annotate('Steve Jobs death', xy=(steves_death, aapl_df['Normalized Close'][steves_death]),
           xytext=(datetime64('2010-01-01'), 3), arrowprops=dict(facecolor='black', shrink=1))
ax.set_xlabel('Date')
ax.set_ylabel('Price')
ax.legend()
ax.set_title('Apple vs Microsoft vs Google Stock Prices');
```

```python
aapl_df = yf.download('AAPL', period='max')
msft_df = yf.download('MSFT', period='max')
goog_df = yf.download('GOOG', period='max')
```

```python
aapl_df.head()
```

```python
msft_df.head()
```

```python
goog_df.head()
```

```python
df = pd.concat([aapl_df['Adj Close'], msft_df['Adj Close'], goog_df['Adj Close']], axis=1, keys=['Apple', 'Microsoft', 'Google'])
df.dropna(how='any', inplace=True)
df.head()
```

```python
df.tail()
```

```python
df.plot(figsize=(10, 6), title='Apple, Microsoft and Google');
```

```python
style.use('seaborn-darkgrid')
fig, axs = plt.subplots(nrows=2, ncols=1, figsize=(10, 12))
df[['Apple', 'Microsoft']].plot(ax=axs[0])
df[['Google']].plot(ax=axs[1])
goog_df[['Close']].plot(ax=axs[1]);
```
