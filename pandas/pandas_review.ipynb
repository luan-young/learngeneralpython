```python
df1 = pd.DataFrame({'Temp':[75, 73, 72, 76],
                    'Humidity':[30,45,32,42],
                    'Precip':[0,0,0,1],
                    'Sample':[1, 2, 3, 4],})

df2 = pd.DataFrame({'Temp':[75, 73, 72, 76],
                    'Wind':[33,35,37,23],
                    'Cloudy':[1, 0, 1, 1],
                    'Sample':[1, 2, 3, 4],})

df1.set_index('Sample', inplace=True)
```

```python
pd.merge(df1,df2, left_index=True, right_on='Sample', how='inner') # INDEX IS LOST
```

```python
pd.merge(df1,df2, on='Sample', how='inner') # INDEX IS LOST AS WELL
```

```python
df2.set_index('Sample', inplace=True)
pd.merge(df1,df2, on='Sample', how='inner') # INDEX PRESERVED
```
